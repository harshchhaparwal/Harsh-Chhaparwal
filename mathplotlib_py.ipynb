```python
#Bar Plot Graph
import matplotlib.pyplot as plt
x=["Python","C++","C","Java"]
y=[22,11,33,44]
plt.xlabel("language")
plt.ylabel("No.")
plt.title("WSCube")
plt.bar(x,y)
plt.show()
```

```python
#Scatter Graph
day=[1,2,3,4,5,6,7]
no=[5,2,4,2,1,4,5]
colors=["red","green","blue","yellow","pink","black","orange"]
sizes=[100,200,300,400,100,200,300]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.5,marker="*")
plt.title("WSCube")
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
#Histogram Graph
no=[49,45,56,16,34,28,33,18,48,54,21,53,43,30,23,48,33,54,47,22,34,59,55,44,54,35,18,24,37,58,16,36,29,43,49,35,55,34,45,55,49,48,14,18,16,58,52,25]
l=[10,20,30,40,50,60]
plt.hist(no,color="b",edgecolor="r",bottom=10)
plt.title("WSCube")
plt.xlabel("No.")
plt.ylabel("Frequency")
plt.show()
```

```python
#Pie Plot Chart
x=[10,20,30,40]
y=["C++","C","Java","Python"]
ex=[0,0.1,0,0]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,shadow=True,startangle=90)
plt.title("WSCube")
plt.show()
```

```python
#Stem Plot
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]
plt.stem(x,y)
plt.title("WSCube")
plt.show()
```

```python
#Box Vs Whisker Plot
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["Python"],showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="y"))
plt.title("WSCube")
plt.show()
```

```python
#Area vs Stack Plot
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]
plt.stackplot(x,area1,area2,area3,labels=["Python","C++","Java"])
plt.title("WSCube")
plt.legend()
plt.show()
```

```python
#Step Plot
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]
plt.step(x,y,where="post")
plt.title("WSCube")
plt.show()
```

```python
#Fill_Between Plot
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]
plt.fill_between(x,y,color="r")
plt.title("WSCube")
plt.show()
```

```python
#SubPlot
x=[1,2,3,4]
y=[1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")
plt.subplot(2,2,2)
plt.pie([1],colors="r")
x1=[10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x)
x2=["a","s","d","f"]
y2=[10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x2,y2)
plt.show()
```
